```python
import yfinance as yf
import pandas as pd
from datetime import datetime
```

```python
aapl = yf.Ticker('AAPL')
```

```python
fin = aapl.financials
```

```python
temp = fin
temp['time'] = None
```

```python
for index,row in temp.iterrows():
    row['time'] = str(datetime.now())
```

```python
temp = temp.reset_index()
temp = temp.set_index('time')
```
